```python
# Change interpreter to global 64-bit
# importing modules
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('income.csv')
df 
```

```python
df.head()
```

```python
df.columns
```

```python
plt.xlabel('Year')
plt.ylabel('Income pc (US$)')
plt.scatter(df.year,df.income,color='red',marker='+')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(df[['year']],df.income,test_size=0.2)
```

```python
X_test
```

```python
X_train
```

```python
# Creating model
model = linear_model.LinearRegression()
model.fit(X_train,y_train)
```

```python
# Accuracy of model
y_predict = model.predict(X_test)
```

```python
model.score(X_test,y_test)
# 75.12%
```

```python
#Income at 2020
model.predict([[2020]])
```

```python
plt.xlabel('Year')
plt.ylabel('Income pc (US$)')
plt.scatter(df.year,df.income,color='red',marker='+')
plt.plot(df.year,model.predict(df[['year']]),color='blue')
```
